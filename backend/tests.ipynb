```python
from collections import defaultdict
from datetime import datetime
import calendar

def calculate_monthly_mean(data):
    # Create a defaultdict to store measurements for each month for each year
    yearly_monthly_data = defaultdict(lambda: defaultdict(list))

    # Iterate through the data and add measurements to the corresponding month and year
    for date_str, value in data.items():
        date = datetime.strptime(date_str, "%Y-%m-%d")
        year = date.year
        month = date.month
        yearly_monthly_data[year][month].append(value)

    mean_data = {}
    for year, monthly_data in yearly_monthly_data.items():
        mean_data[year] = {}
        for month, values in monthly_data.items():
            month_name = calendar.month_name[month]
            mean_data[year][month_name] = sum(values) / len(values)

    return mean_data
```

```python
# predict the water quality index based on a set of measurements
def predict_wqi(values):
    import joblib
    model = joblib.load('regressor_model.pkl')
    if(len(values.shape) == 1):
        values = values.reshape(1,-1)
    
    y_pred = model.predict(values)
    
    return y_pred


# determine the water quality class based on a wqi value
# the function takes an array
def get_wqc(wqi):
    import numpy as np
    wqc = []
    wqi = np.array(wqi)
    if(len(wqi.shape) == 0):
        wqi = wqi.reshape(1,-1)
        
    print(wqi)
        
    for i in range(wqi.shape[0]):
        if(wqi[i] <= 50):
            wqc.append("good")
        elif(wqi[i] <= 100):
            wqc.append("poor")
        else:
            wqc.append("unsuitable")
    return np.array(wqc)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.stats import zscore

def wqi_gauge(specified_date):
    df = data.copy()
    print(df)
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)
    index_of_date = df.index.get_loc(specified_date)
    
    df = df.drop(columns=["STATION CODE", "LOCATIONS", "STATE", "Temp", "year", "date"])
    
    for column in df.columns:
        df[column] = pd.to_numeric(df[column], errors='coerce')
        
    df.fillna(df.median(), inplace=True)
    
    # Checking for outliers in the full dfset
    z_scores = df.apply(zscore)
    outliers_count = (z_scores.abs() > 3).sum()
    df_outliers = np.where(np.abs(z_scores) > 3)[0]
    print(outliers_count)
    df_filtered = df.drop(df.index[df_outliers])
    
    # Scale the df
    scaler = StandardScaler()
    X = df_filtered.values
    X_scaled = scaler.fit_transform(X)
    
    return X_scaled[index_of_date]
```
